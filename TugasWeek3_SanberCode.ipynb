# Tugas Hari KEDUA_CUSTOMIZING VISUALIZATION

<h2>Soal 1:</h2>

Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
# code here
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(12,7))
ax.plot(x, cos, color='r', label='Cosx')
ax.plot(x, cos3, linestyle='--', color='g', label='3 Cosx')
ax.plot(x, sin, linestyle='-.', color='b', label='Sin(x-90)')
ax.plot(x,sin2, marker='X', linestyle='None', color='y', label='2 Sinx')

ax.set_xlabel('X-Axis Label')
ax.set_ylabel('Y-Axis Label')
ax.set_title('Grafik Cos Sin')
plt.legend(loc='center')
```

# Tugas Hari KETIGA_MULTIPLE SUBPLOTS

```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# code here
fig,ax = plt.subplots(nrows=1,ncols=3,sharex=True,sharey=True,figsize=(15,8))

ax[0].plot(x,tan, label='tanX')
ax[0].set_title('Grafik tanx')

ax[1].plot(x,sin,color='red', label = 'sinX')
ax[1].set_title('Grafik sinx')

ax[2].plot(x,cos,color='green', label = 'cosX')
ax[2].set_title('Grafik cosx')

fig.legend(loc='upper center')
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
# code here
fig, ax = plt.subplots(figsize=(15,8))
ax.plot(x, y)
ax.set_title('Grafik CosX')

graf1 = ax.inset_axes([0.9, 0.9, 0.3, 0.3])
graf1.plot(x, y2, color='grey')
graf1.plot(x, y, color='blue')
graf1.set_title('Grafik CosX^2')

graf2 = ax.inset_axes([0.9, -0.1, 0.3, 0.3])
graf2.plot(x, y2, color='grey')
graf2.plot(x, y, color='blue')
graf2.set_title('Grafik CosX^3')

graf3 = ax.inset_axes([-0.2, -0.1, 0.3, 0.3])
graf3.plot(x, y4, color='red')
graf2.plot(x, y, color='blue')
graf3.set_title('Grafik CosX^4')

graf4 = ax.inset_axes([-0.2, 0.9, 0.3, 0.3])
graf4.plot(x, y5, color='yellow')
graf4.plot(x, y, color='blue')
graf4.set_title('Grafik CosX^5')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('austin_weather.csv')
df.head()
```

```python
df.info()
```

<h2>Soal 1: Visualisasi Scatter Plot Perbandingan Kuantitatif</h2>

Pada tugas kali ini kita akan mengamati nilai DewPointAvg (F) dengan mengamati nilai  HumidityAvg (%), TempAvg (F), dan WindAvg (MPG)

Perhatikan bahwa data kita tidaklah siap untuk di analisis, salah satunya tipe data dari DewPointAvg (F), HumidityAvg (%), dan WindAvg (MPG) adalah object, padahalnya data nya ber isi numeric. maka :
- Ubahlah tipe data tersebut menjadi tipe data float

Kemudian: 

- Kalian tidak akan dengan mudah mengubah tipe data tersebut karena column tersebut mempunyai nilai '-' yang dimana tidak bisa di ubah ke bentuk float, maka replace lah terlebih dahulu data yang bernilai '-' dengan nilai NaN, gunakan method .replace(). baca dokumentasi https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.replace.html
- Isi nilai nan dengan nilai sebelumnya di row tersebut. gunakan method .fillna() dengan argument method bernilai 'ffill', baca dokumentasi https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.fillna.html
- Sekarang ubah tipe datanya dengan float, gunakan method .astype(), baca dokumentasi https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.astype.html

Setelah ini sebagian data siap untuk di jadikan bahan analisis. maka:

Buahlah visualisasi perbandingan kuantitatif scatter plot, sehingga menghasilkan gambar seperti dibawah:

ket: 

- colormap adalah 'coolwarm'
- berikat warna terhadap setiap data poin dengan nilai dari column TempAvgF
- berikan size terhadap setiap data poin dengan nilai dari column WindAvgMPH, kalikan dengan 20 agar size terlihat lebih besar


Berikan pendapat dari insight yang bisa di dapat dari visualisasi perbandingan kuantitatif ini!!!

![alt text](https://drive.google.com/uc?id=1Pp5oIc-39mcw2vrw6xgCRM9GsvyRtPI5)

```python
#menghitung jumlah nan
df.isnull().sum()
#mengubah tipe data menjadi float
df.replace('-',np.nan,inplace=True)
df.fillna(method='ffill',inplace=True)
df=df[['TempAvgF','WindAvgMPH','HumidityAvgPercent','DewPointAvgF']].astype(float)
```

```python
df.info()
```

```python
#membuat scatter plot
plt.figure(figsize=(14,8))

plt.scatter(x=df['HumidityAvgPercent'],y=df['DewPointAvgF'],s=df['WindAvgMPH']*20,c=df['TempAvgF'],alpha=0.7,cmap='coolwarm')

plt.xlabel('Humidity Avg %')
plt.ylabel('Dew Point Avg (F)')
plt.title('Austin Weather')
plt.colorbar()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('vgsales.csv')
df.head()
```

<h2>Soal 2: Preparing Data</h2>

Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk setiap Region Kecuali Global_Sales berdasarkan kategori Genre.

```python
#code here
import warnings
warnings.filterwarnings('ignore')
df_group = df.groupby('Genre')['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales'].mean()
df_group
```

<h2>Soal 3: Perbandingan Kuantitatif Barplot - Grouping Visualisasi Dengan Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menghasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
for i,j in enumerate(df_group.columns):
  print(i,j)
```

```python
#code here
plt.style.use('ggplot')
x = np.arange(0, 20, 5)
fig, ax = plt.subplots(figsize=(9,5))
wid = 0.3

i, j = 0, 1
while i < len(df_group.index) :
    ax.bar(x+j, df_group.iloc[i], width=wid, label=df_group.index[i])
    i +=1
    j +=0.3
    
ax.set_xticklabels([ '', '', 'NA_Sales', '', 'EU_Sales', '', 'JP_Sales', '', 'Other_Sales'])
ax.set_title('Mean Sales Video Games by Genre')
ax.set_xlabel('Region Sales')
ax.set_ylabel('Mean Sales')
fig.legend()
plt.show()
```

Jawab: 
    Berdasarkan hasil dari perbandingan barplot tersebut diperoleh :
Rata-rata penjualan NA_sales paling tinggi yaitu platform
Rata-rata penjualan EU_Sales paling tinggi yaitu shooter
Rata-rata penjualan JP_sales paling tinggi yaitu actiom
Rata-rata penjualan Other_sales paling tinggi yaitu shooter

 Dan dapat disimpulkan bahwa yang memiliki rata-rata penjualan paling banyak yaitu NA_Sales.

<h2>Soal 4: Perbandingan Kuantitatif Barplot - Stack Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menhasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
#code here
#code here
plt.style.use("ggplot")
fig, ax=plt.subplots(figsize=(17,8))
x = df_group.columns
width =0.6
y_bottom=0

for row in df_group.iterrows():
  plt.bar(x, row[1], width, label= row[0],bottom= y_bottom)
  y_bottom += row[1]

ax.set_title('Mean Sales Video Game by Genre')
ax.set_xlabel('Region Sales')
ax.set_ylabel('Mean Sales')
plt.legend()
plt.show()
```

# Tugas Hari KELIMA_SEABORN

```python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df = sns.load_dataset('diamonds')
df
```

# Soal 1: Scatter Plot

Buatlah scatterplot menggunakan Seaborn dari data df di atas, variasikan menggunakan palette yang telah disediakan sehingga didapatkan output sesuai expected.

```python
import warnings
warnings.filterwarnings('ignore')
sns.set({'figure.figsize':(13,13)})
palette = "ch:r=-.2,d=.3_r"

#code here
x = df['carat']
y = df['price']
clarity = df['clarity']
depth = df['depth']

sns.scatterplot(x, y, hue = clarity, size = depth, palette = palette, sizes = (5,10))
```

Expected Output:

![alt text](https://drive.google.com/uc?id=1JLtkSpx0TfSF7fMSBcZuXaY_tNkR5LYc)

# Soal 2: Catplot

Buatlah sebuah Categorical plot dengan kind = 'boxplot', sehingga diperoleh hasil seperti expected output berikut:


```python
import warnings
warnings.filterwarnings('ignore')

#code here
sns.set({'figure.figsize':(13,13)})
with sns.axes_style(style='ticks'):
    cat = sns.catplot('clarity', 'price', col = 'cut', data = df, kind = 'box',  showfliers=False)
    cat.set_axis_labels('clarity', 'price');
```

# Soal 3: Pair Plot

Ambil 300 data secara acak dari df, kemudian buat pair plotnya dari 300 data tersebut untuk kolom carat, depth, table, dan price serta markers =  ['p', 'o', 's', 'D', 'X'] sehingga dihasilkan output sesuai expected.

```python
df = df.sample(n=300, random_state=123)

sns.pairplot(df, hue='cut', corner=True,vars=['carat','depth','table','price'],markers = ['p', 'o', 's', 'D', 'X'])
```
